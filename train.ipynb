```python
import numpy as np

T_l, T_g = 140, 960000
seq1 = np.random.uniform(-5, 5, size = T_l) + 20
seq2 = np.random.uniform(-5, 5, size = T_l) + 20
```

```python
import matplotlib.pyplot as plt
import numpy as np
from population import Population

npop = 100
num_days = 10000
p_cross = 0.9
p_mute = 1e-5


stock_generator = Population.alternate_sequence_generator(seq1 = seq1, seq2 = seq2, T_global=T_g)

pop = Population(npop=npop, p_cross=p_cross, p_mute=p_mute, stock_generator=stock_generator, num_elites=0, t_ga=T_l)
pop.run_population(num_days=num_days)

fig, ax = plt.subplots(nrows=3, ncols=1)
ax[0].plot(pop.wealthiest_history, 'ok-', markersize=2)
ax[2].plot(pop.prices, 'ok-', markersize=1)
ax[1].plot(pop.mean_wealth_history, 'ok-', markersize=2)
# ax[0].set_ylim(0,500)
# ax[1].set_ylim(0,500)
ax[2].set_ylim(10, 30)
ax[0].grid()
ax[1].grid()
ax[2].grid()

print(f'All time wealth: {sum(pop.mean_wealth_history)}')
```
